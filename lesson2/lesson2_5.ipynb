提示：

平方使用 ** 運算子：height ** 2

這題解題步驟與程式碼示例如下：

建立變數 weight 儲存體重 72 公斤

建立變數 height 儲存身高 175 公分，並除以 100 轉換成公尺

使用 ** 運算子計算身高平方

計算 BMI，公式為 體重除以身高平方

使用 round() 函式將 BMI 四捨五入到小數點後 2 位

印出結果



```python
def calculate_bmi(weight, height_cm):
    height_m = height_cm / 100
    bmi = weight / (height_m ** 2)
    return round(bmi, 2)

def bmi_category(bmi):
    if bmi < 18.5:
        return "過輕"
    elif 18.5 <= bmi < 24:
        return "正常"
    elif 24 <= bmi < 27:
        return "過重"
    elif 27 <= bmi < 30:
        return "輕度肥胖"
    elif 30 <= bmi < 35:
        return "中度肥胖"
    else:
        return "重度肥胖"

def get_positive_float(prompt):
    while True:
        try:
            value = float(input(prompt))
            if value <= 0:
                print("請輸入正數。")
                continue
            return value
        except ValueError:
            print("輸入錯誤，請輸入數字。")

def main():
    weight = get_positive_float("請輸入體重（公斤）：")
    height = get_positive_float("請輸入身高（公分）：")
    bmi = calculate_bmi(weight, height)
    category = bmi_category(bmi)
    print(f"BMI值為: {bmi}，分類為: {category}")

if __name__ == "__main__":
    main()
```

```python
import unittest

class TestBMICalculator(unittest.TestCase):
    def test_calculate_bmi(self):
        self.assertAlmostEqual(calculate_bmi(72, 175), 23.51)
    
    def test_bmi_category(self):
        self.assertEqual(bmi_category(17), "過輕")
        self.assertEqual(bmi_category(22), "正常")
        self.assertEqual(bmi_category(25), "過重")
        self.assertEqual(bmi_category(28), "輕度肥胖")
        self.assertEqual(bmi_category(32), "中度肥胖")
        self.assertEqual(bmi_category(36), "重度肥胖")

if __name__ == "__main__":
    unittest.main()
```
